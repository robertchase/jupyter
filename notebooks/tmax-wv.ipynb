```python
import pandas as pd
import seaborn as sns
from datetime import date
```

```python
data = pd.read_csv("data/2805257.csv")
```

```python
data.head()
```

```python
data.groupby("STATION")[["LATITUDE","LONGITUDE"]]
```

```python
data["date"] = pd.to_datetime(data["DATE"]).apply(lambda x: x.date())
```

```python
data["basedate"] = (pd.to_datetime(data["DATE"]) - pd.to_datetime('1880-01-01')).apply(lambda x: x.days)
```

```python
s1 = data[data["STATION"] == "USC00441593"]
s2 = data[data["STATION"] == "USC00441598"]
s3 = data[data["STATION"] == "USC00444768"]
s4 = data[data["STATION"] == "USW00013734"]
```

```python
sns.lineplot(data=s1, x="date", y="TMAX")
```

```python
sns.regplot(data=s1,x="basedate",y="TMAX",scatter=False)
```

```python
sns.lineplot(data=s2, x="date", y="TMAX")
```

```python
sns.regplot(data=s2,x="basedate",y="TMAX",scatter=False)
```

```python
sns.lineplot(data=s3, x="date", y="TMAX")
```

```python
s3 = s3[s3["date"] > date(1950,1,1)]
```

```python
sns.lineplot(data=s3, x="date", y="TMAX")
```

```python
sns.regplot(data=s3,x="basedate",y="TMAX",scatter=False)
```

```python
sns.lineplot(data=s4, x="date", y="TMAX")
```

```python
sns.regplot(data=s4,x="basedate",y="TMAX",scatter=False)
```
