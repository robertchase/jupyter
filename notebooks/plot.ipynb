```python
from scipy import special, optimize, linspace
f = lambda x: -special.jv(3, x)
sol = optimize.minimize(f, 1.0)
x = linspace(0, 10, 5000)
x
```

```python
from matplotlib.pyplot import plot
plot(x, special.jv(3, x), '-', sol.x, -sol.fun, 'o')
```
